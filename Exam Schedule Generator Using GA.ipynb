```python
import numpy as np
import pandas as pd

# Constants
ROOMS = ['C301', 'C302', 'C303', 'C304', 'C305', 'C306', 'C307', 'C308', 'C309', 'C310']
DAYS = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
TIME_SLOTS = [(9, 11), (11, 13), (13, 15), (15, 17)]  # Exam start times and durations
POPULATION_SIZE = 50
NUM_GENERATIONS = 100
MUTATION_RATE = 0.01

# Load data
teachers_df = pd.read_csv(r"C:\Users\Pak\Downloads\Dataset\teachers.csv")
final_df = pd.read_csv(r"C:\Users\Pak\Downloads\Dataset\final_modified.csv")

# Extract unique courses
courses = pd.unique(final_df[['Course1', 'Course2', 'Course3']].values.ravel('K'))
courses = [course for course in courses if pd.notna(course)]  # Remove NaN entries

# Function to create an initial population with valid schedules
def create_initial_population():
    population = []
    for _ in range(POPULATION_SIZE):
        chromosome = generate_valid_chromosome()
        population.append(chromosome)
    return population

# Generate a chromosome that satisfies all hard constraints
def generate_valid_chromosome():
    chromosome = []
    for course in courses:
        valid = False
        while not valid:
            room = np.random.choice(ROOMS)
            day = np.random.choice(DAYS)
            time_slot_index = np.random.randint(len(TIME_SLOTS))
            time_slot = TIME_SLOTS[time_slot_index]
            teacher = np.random.choice(teachers_df['Names'].tolist())
            gene = (course, room, day, time_slot[0], teacher)
            if validate_gene(gene, chromosome):
                chromosome.append(gene)
                valid = True
    return chromosome

# Validate if adding the new gene violates any hard constraints
def validate_gene(new_gene, chromosome):
    course, room, day, start_time, teacher = new_gene
    for gene in chromosome:
        _, existing_room, existing_day, existing_start, existing_teacher = gene
        if day == existing_day:
            if teacher == existing_teacher and start_time == existing_start:
                return False  # Teacher cannot be in two places at once
    return True

# Function to calculate fitness
def calculate_fitness(chromosome):
    fitness = 100  # Start with a perfect score and deduct for constraint violations
    for gene in chromosome:
        course, room, day, start_time, teacher = gene
        # Example of hard constraint check:
        if day == 'Friday' and (11 <= start_time < 13):
            fitness -= 5  # Penalize exams scheduled during the Friday break
    return fitness

# Roulette wheel selection
import random

def roulette_wheel_selection(population, fitnesses):
    total_fitness = sum(fitnesses)
    selection_probs = [f / total_fitness for f in fitnesses]
    return random.choices(population, weights=selection_probs, k=POPULATION_SIZE)

def genetic_algorithm():
    population = create_initial_population()
    best_fitness_over_time = []

    for generation in range(NUM_GENERATIONS):
        fitnesses = [calculate_fitness(chromosome) for chromosome in population]
        best_fitness_over_time.append(max(fitnesses))
        new_population = roulette_wheel_selection(population, fitnesses)
        # Implement crossover and mutation here if needed
        population = new_population

        print(f'Generation {generation+1}, Best Fitness: {max(fitnesses)}')

    return population, best_fitness_over_time

# Run the genetic algorithm
best_population, fitness_over_time = genetic_algorithm()
display_results(best_population, fitness_over_time)
```

```python
def roulette_wheel_selection(population, num_parents):
    # Calculate fitness for each chromosome in the population
    fitness_scores = [calculate_fitness(chromosome) for chromosome in population]
    total_fitness = sum(fitness_scores)
    
    # Calculate selection probabilities
    selection_probabilities = [fitness / total_fitness for fitness in fitness_scores]
    
    # Select parents using the roulette wheel method
    selected_indices = np.random.choice(len(population), size=num_parents, p=selection_probabilities)
    selected_parents = [population[i] for i in selected_indices]
    
    return selected_parents

# Apply roulette wheel selection to select 2 parents from the corrected initial population
selected_parents = roulette_wheel_selection(initial_population_corrected, 2)
len(selected_parents), selected_parents[0][:3]  # Display the first 3 genes of the first selected parent
```

```python
def single_point_crossover(parent1, parent2):
    # Ensure the parents are of the same length
    length = min(len(parent1), len(parent2))
    if length == 0:
        return [], []

    # Randomly choose a crossover point
    crossover_point = random.randint(1, length - 1)

    # Create offspring by swapping genes past the crossover point
    offspring1 = parent1[:crossover_point] + parent2[crossover_point:]
    offspring2 = parent2[:crossover_point] + parent1[crossover_point:]
    
    return offspring1, offspring2

# Perform single-point crossover on the selected parents
offspring1, offspring2 = single_point_crossover(selected_parents[0], selected_parents[1])
offspring1[:3], offspring2[:3]  # Display the first 3 genes of each offspring
```

```python
def mutate_chromosome(chromosome, mutation_rate=0.01):
    # Iterate through each gene in the chromosome
    for i in range(len(chromosome)):
        if random.random() < mutation_rate:  # Apply mutation with a probability of mutation_rate
            # Randomly choose an aspect to mutate: Classroom, Day, or Start Hour
            mutation_choice = random.choice(['Classroom', 'Day', 'Start Hour'])
            
            if mutation_choice == 'Classroom':
                # Change to a different classroom
                new_classroom = random.choice(classrooms)
                chromosome[i]['Classroom'] = new_classroom
            
            elif mutation_choice == 'Day':
                # Change to a different day
                new_day = random.choice(days)
                chromosome[i]['Day'] = new_day
            
            elif mutation_choice == 'Start Hour':
                # Change to a different start hour
                new_hour = f"{random.choice(hours)}:00"
                chromosome[i]['Start Hour'] = new_hour
                # Update end hour assuming 1-hour exams
                start_time = int(new_hour.split(':')[0])
                chromosome[i]['End Hour'] = f"{start_time + 1}:00"
    
    return chromosome

# Apply mutation to the first offspring
mutated_offspring = mutate_chromosome(offspring1)
mutated_offspring[:3]  # Display the first 3 genes of the mutated offspring
```

```python
def run_genetic_algorithm(num_generations=50, population_size=20, num_parents=10):
    # Initialize the population
    population = generate_initial_population_corrected(population_size)
    
    # Track fitness history for analysis
    fitness_history = []

    for generation in range(num_generations):
        # Evaluate the fitness of the current population
        fitness_scores = [calculate_fitness(chromosome) for chromosome in population]
        average_fitness = sum(fitness_scores) / len(fitness_scores)
        fitness_history.append(average_fitness)
        
        # Print average fitness of the current generation
        print(f"Generation {generation + 1}: Average Fitness = {average_fitness:.2f}")
        
        # Selection: Select parents for the next generation
        parents = roulette_wheel_selection(population, num_parents)
        
        # Crossover and Mutation: Generate new offspring for the next generation
        new_population = []
        while len(new_population) < population_size:
            # Randomly select two parents for crossover
            parent1, parent2 = random.sample(parents, 2)
            offspring1, offspring2 = single_point_crossover(parent1, parent2)
            # Mutate the offspring
            mutated_offspring1 = mutate_chromosome(offspring1)
            mutated_offspring2 = mutate_chromosome(offspring2)
            # Add the new offspring to the population
            new_population.extend([mutated_offspring1, mutated_offspring2])
        
        # Replace the old population with the new population
        population = new_population[:population_size]
    
    return fitness_history

# Run the genetic algorithm
fitness_history = run_genetic_algorithm(num_generations=10, population_size=10, num_parents=5)
```
